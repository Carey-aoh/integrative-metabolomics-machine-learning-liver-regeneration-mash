```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white")       
df_pos = pd.read_csv("pos.csv", index_col=0)
df_neg = pd.read_csv("neg.csv", index_col=0)
r2_pos = df_pos.T.corr(method="pearson") ** 2
r2_neg = df_neg.T.corr(method="pearson") ** 2
fig, (ax1, ax2) = plt.subplots(
    1, 2, figsize=(14, 6), sharey=False, dpi=120
)
vmin, vmax = 0.96, 1.00
cbar_opts  = dict(label="R²", shrink=0.80)
sns.heatmap(
    r2_pos, ax=ax1, cmap="Blues", vmin=vmin, vmax=vmax,
    annot=True, fmt=".3f", square=True,
    linewidths=0.5, linecolor="white",
    cbar_kws=cbar_opts
)
ax1.set_title("Pearson correlation between pos QC samples", pad=12)
sns.heatmap(
    r2_neg, ax=ax2, cmap="Blues", vmin=vmin, vmax=vmax,
    annot=True, fmt=".3f", square=True,
    linewidths=0.5, linecolor="white",
    cbar_kws=cbar_opts
)
ax2.set_title("Pearson correlation between neg QC samples", pad=12)
plt.tight_layout(rect=[0.04, 0, 1, 0.96])
plt.savefig("QC_correlation_heatmap.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
